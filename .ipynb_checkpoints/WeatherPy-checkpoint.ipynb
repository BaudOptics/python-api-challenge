```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
from scipy import stats
from citipy import citipy
from config import wkey
```

```python
#making a dataframe of random numbers between 0 and 1
lat_lon = pd.DataFrame(np.random.rand(500, 2))
lat = lat_lon[0]
lon = lat_lon[1]

over90 = 0
over180 = 0

for i in range(len(lat_lon)):
    #multiplying latitudes by random numbers between -100 and 100
    lat_lon[0][i] = lat_lon[0][i]*np.random.uniform(-100,100)
    #constraining latitudes to -90 and 90
    if lat_lon[0][i] >90:
        over90 = lat_lon[0][i] - 90
        lat_lon[0][i] = 90 - over90
    elif lat_lon[0][i] <-90:
        over90 = lat_lon[0][i] + 90
        lat_lon[0][i] = -90 + over90
        
    #doing the same for longitude but using -180 and 180
    lat_lon[1][i] = lat_lon[1][i]*np.random.uniform(-180,180)
    #constraining latitudes to -180 and 180
    if lat_lon[1][i] >180:
        over180 = lat_lon[1][i] - 180
        lat_lon[1][i] = 180 - over180
    elif lat_lon[1][i] <-180:
        over180 = lat_lon[1][i] + 180
        lat_lon[1][i] = -180 + over180

lat_lon.columns = ["Latitude","Longitude"]
lat_lon
```

```python
#getting city names from dataframe of lat and lng
rand_cities = []
for i in range(len(lat_lon)):
    rand_city = citipy.nearest_city(lat_lon["Latitude"][i],lat_lon["Longitude"][i])
    rand_cities.append(rand_city.city_name)
rand_cities
```

```python
#adding cities to the lat_lon dataframe
lat_lon["Cities"] = rand_cities
cities = lat_lon
cities.drop_duplicates(subset=["Cities"], inplace=True)
cities
```

```python
while len(cities)<500:
    new_lat_lon = pd.DataFrame(np.random.rand(1, 2))
    for i in range(len(new_lat_lon)):
        #multiplying latitudes by random numbers between -100 and 100
        new_lat_lon[0][i] = new_lat_lon[0][i]*np.random.uniform(-100,100)
        #constraining latitudes to -90 and 90
        if new_lat_lon[0][i] >90:
            over90 = new_lat_lon[0][i] - 90
            new_lat_lon[0][i] = 90 - over90
        elif new_lat_lon[0][i] <-90:
            over90 = new_lat_lon[0][i] + 90
            new_lat_lon[0][i] = -90 + over90

        #doing the same for longitude but using -180 and 180
        new_lat_lon[1][i] = new_lat_lon[1][i]*np.random.uniform(-180,180)
        #constraining latitudes to -180 and 180
        if new_lat_lon[1][i] >180:
            over180 = new_lat_lon[1][i] - 180
            new_lat_lon[1][i] = 180 - over180
        elif new_lat_lon[1][i] <-180:
            over180 = new_lat_lon[1][i] + 180
            new_lat_lon[1][i] = -180 + over180
    new_lat_lon.columns = ["Latitude","Longitude"]
    new_city = citipy.nearest_city(new_lat_lon["Latitude"][i],new_lat_lon["Longitude"][i])
    new_lat_lon["Cities"] = new_city.city_name
    if new_lat_lon["Cities"][0] not in cities["Cities"]:
        cities = cities.append(new_lat_lon)
    cities.drop_duplicates(subset="Cities",inplace=True)
cities.drop_duplicates(subset="Cities",inplace=True)
cities.reset_index(drop=True,inplace=True)
cities
```

```python
#configuring the URL call
url = "http://api.openweathermap.org/data/2.5/weather?"
city = "london"
units = "imperial"

# Build query URL
query_url = url + "appid=" + wkey + "&q=" + city +"&units=" + units

weather_response = requests.get(query_url)
weather_json = weather_response.json()
#temp
#humidity
#cloudiness
#wind speed

weather_json
```

```python
weather_df = pd.DataFrame()
for i in range(len(cities)):
    city = cities["Cities"][i]
    query_url = url + "appid=" + wkey + "&q=" + city +"&units=" + units
    weather_response = requests.get(query_url)

    weather_json = weather_response.json()
    if weather_json['cod'] == 200:
        weather_info={"Temperature":weather_json['main']['temp'],
                      "Humidity":weather_json['main']['humidity'],
                      "Cloudiness":weather_json['clouds']['all'],
                      "Wind Speed":weather_json['wind']['speed']}
        cities['Latitude'][i] = weather_json["coord"]['lat']
        cities['Longitude'][i] = weather_json['coord']['lon']
        weather_df = weather_df.append(weather_info,ignore_index=True)
    elif weather_json['cod'] == "404":
        weather_info={"Temperature":np.nan,
                      "Humidity":np.nan,
                      "Cloudiness":np.nan,
                      "Wind Speed":np.nan}
        weather_df = weather_df.append(weather_info,ignore_index=True)
weather_df
```

```python
merged_df = pd.merge(cities,weather_df,left_index=True, right_index=True)

merged_df = merged_df.dropna()
merged_df.to_csv("Weather_Info.csv")
merged_df
```

```python
north = merged_df[merged_df["Latitude"]>=0]
south = merged_df[merged_df["Latitude"]<=0]
```

```python
lat = north.iloc[:,0]
temp = north.iloc[:,5]
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(lat, temp)
regress_values = lat * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(lat,temp)
plt.plot(lat,regress_values,"r-")

plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel("Temperature")
plt.ylabel("Latitude")
plt.show()
```
